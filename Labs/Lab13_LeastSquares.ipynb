## Part 1: write at least **four** implementations to solve the least-squares problem



```python
# Problem 0: use this for debugging
problem_label = 'Problem 0'
rng     = np.random.default_rng(12345)
n       = 20
m       = 40
A       = rng.standard_normal( (m,n), dtype=np.double )
xTrue   = rng.standard_normal(n)
b       = A@xTrue
print(f'Matrix is size {m} x {n}')
print(f'Condition number of A is {np.linalg.cond(A):.2e}')
```

```python
# Your solutions


## Solving methods
def ls_normal_eq(A, b): # Least squares via normal equations
    from scipy.linalg import solve
    ATA = A.T @ A
    ATb = A.T @ b
    x = solve(ATA, ATb, assume_a='pos')  # SPD system
    return x

def ls_qr(A, b): # least squares via QR factorization
    from scipy.linalg import qr, solve_triangular
    Q, R = qr(A, mode='economic')  # reduced QR
    Qt_b = Q.T @ b
    x = solve_triangular(R, Qt_b)
    return x

def ls_svd(A, b): # least squares via SVD
    from scipy.linalg import svd
    U, s, Vt = svd(A, full_matrices=False)
    c = U.T @ b
    w = c / s
    x = Vt.T @ w
    return x

def ls_lstsq(A, b): # Use built in solver
    from scipy.linalg import lstsq
    x, *_ = lstsq(A, b)
    return x


## Error metric
def linf_error(x_true, x_est):
    return np.max(np.abs(x_true - x_est))

## Run experiemnt

def time_method(func, A, b, x_true):
    start_wall = time.time()
    start_cpu = time.process_time()

    x_est = func(A, b)

    end_cpu = time.process_time()
    end_wall = time.time()

    return {
        "x_est": x_est,
        "wall_time": end_wall - start_wall,
        "cpu_time": end_cpu - start_cpu,
        "linf_error": linf_error(x_true, x_est)
    }

def run_experiment(problem_label, A, xTrue, b):
    print(f"\n=== {problem_label} ===")
    print(f"Matrix shape: {A.shape}")
    cond_num = np.linalg.cond(A)
    print(f"Condition number: {cond_num:.2e}\n")

    methods = {
        "Normal Eq": ls_normal_eq,
        "QR": ls_qr,
        "SVD": ls_svd,
        "lstsq": ls_lstsq
    }

    results = {}

    for name, func in methods.items():
        print(f"Running {name}...")
        res = time_method(func, A, b, xTrue)
        results[name] = res

    # Print results table
    print("\nResults:")
    print(f"{'Method':10s} | {'Wall Time (s)':>12s} | {'CPU Time (s)':>12s} | {'Linf Error':>12s}")
    print("-"*60)

    for name, res in results.items():
        print(f"{name:10s} | {res['wall_time']:12.4f} | {res['cpu_time']:12.4f} | {res['linf_error']:12.2e}")

    return results

results0 = run_experiment("Problem 0", A, xTrue, b)
```

On problem 0, which is a well conditioned system, all the methods are very accurate with errors near machine precision, confirming correctness of the implementations. Differences in performance and stability are not significant for the small and well-behaved case.

## Part 2: Try solving some larger problems
- Record the **error** in the infinity norm, and...
- Record the **time** it takes

To measure time, in jupyter/colab/iPython, you can use a line like

`%time x = myFunction(...)`

and it will time how long it takes `myFunction(...)` to run


What **observations** can you make?

Note that if you use `%time`, it will return both the:
1. CPU time, which is the sum of how long all cores on the computer spend. So if you have 4 cores, and each core takes 3 seconds (and is fully utilized), the CPU time is 12 seconds.
2. "Wall" time, which refers to "wall clock" time. This is the actual time elapsed from when you start to when you stop.

If the CPU time is 12 seconds and the wall time is 3 seconds, this indicates that the code ran in parallel.

```python
# Problem 1
problem_label = 'Problem 1'
rng     = np.random.default_rng(12345)
n       = int(5e3)
m       = n+10
# To give ourselves an added challenge (and to be fair for when we compare to the GPU),
#   let's do this in single precision rather than double precision, so this means
#   we start with about 8 digits of accuracy
A       = rng.standard_normal( (m,n), dtype=np.single )
xTrue   = rng.standard_normal(n)
b       = A@xTrue
#print(f'Matrix is size {m} x {n}')
#print(f'Condition number of A is {np.linalg.cond(A):.2e}') # takes a bit of time to run! It's about 2e3 if n=5e3

results1 = run_experiment("Problem 1", A, xTrue, b)
```

```python
# Problem 2
problem_label = 'Problem 2'
rng = np.random.default_rng(12345)

n   = int(2e3)
n2  = int(n/2)
m   = int(1e4)
A       = rng.standard_normal( (m,n2) )
c       = 1e-2 # make this smaller, like 1e-4, to make the matrix more ill-conditioned
A       = np.hstack(  (A,A+c*rng.standard_normal( (m,n2) )), dtype=np.single )
xTrue   = rng.standard_normal(n)
b       = A@xTrue
#print(f'Matrix is size {m} x {n}')
#print(f'Condition number of A is {np.linalg.cond(A):.2e}') # takes a bit of time to run! It's about 4.1e2

results2 = run_experiment("Problem 2", A, xTrue, b)
```

```python
print("\n=== Summary ===")
print(f"{'Problem':10s} | {'Method':10s} | {'Wall Time (s)':>12s} | {'CPU Time (s)':>12s} | {'Linf Error':>12s}")
print("-"*70)
for problem_label, results in zip(["Problem 0", "Problem 1", "Problem 2"], [results0, results1, results2]):
    for method_name, res in results.items():
        print(f"{problem_label:10s} | {method_name:10s} | {res['wall_time']:12.4f} | {res['cpu_time']:12.4f} | {res['linf_error']:12.2e}")
    print("-"*70)
```

Across all the problems, the normal equation method was consistently the fastest, but produced more error than the other methods, especially as the problem size increased or the matrix became more ill-conditioned (due to squaring the condition number). 

QR decomposition was a good mix between speed and accuracy, it was much more accurate than the normal equaitons and faster than SVD, making it probably the best choice.

SVD was the most computationally expensive, with much higher wall and CPU times, especially in problem 1. However, its accuracy was consistently high even as the matrix became more ill-conditioned. 

The built in lstsq solver produced the most accurate results across all the problems, but had the highest runtime.
