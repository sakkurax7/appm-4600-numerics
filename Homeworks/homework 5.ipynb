```python
import numpy as np

def michalewicz(x):
    fe = []
    for i, xi in enumerate(x):
        fe.append( -( np.sin(xi)*np.sin((i+1) * xi **2)**20 ) )
    return sum(fe)

m = 1000 # Plenty for 0.1 accuracy
x = np.linspace(0, np.pi, m)
X1, X2 = np.meshgrid(x, x)

Z = michalewicz([X1, X2])

min_value = np.min(Z)
min_index = np.argmin(Z)
coords = np.unravel_index(min_index, Z.shape)

print("Minimum value:", min_value)
print("At x =[", x[coords[0]], x[coords[1]], "]")
```
